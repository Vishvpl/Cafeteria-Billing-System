```python
from tkinter import *
root=Tk()
root.title("Cafeteria")
root.geometry("800x600")
root.config(bg='brown')


heading=Label(text="Welcome To Our Cafeteria",bg="brown",pady=20,padx=20,font=('arial',20),fg='white')
sub_head=Label(text="What would you like to have?",bg='brown',pady=20,padx=20,font=('arial',18),fg='white')

sum=IntVar()
coffee=IntVar()
tea=IntVar()
samosa=IntVar()
momos=IntVar()
burger=IntVar()
cd=IntVar()
sum=0
prevstate={'coffee':False,'tea':False,'momos':False,'burger':False,'samosa':False,'cd':False}
def show():
    global sum
    if coffee.get()==30:
        if not prevstate['coffee']:
            sum+=30
            prevstate['coffee']=True
        # coffee.set(-1)
    else:
        if prevstate['coffee']:
            sum-=30
            prevstate['coffee']=False
    if tea.get()==20:
        if not prevstate['tea']:
            sum+=20
            prevstate['tea']=True

    else:
        if prevstate['tea']:
            sum-=20
            prevstate['tea']=False
    if momos.get()==60:
        if not prevstate['momos']:
            sum+=60
            prevstate['momos']=True
        
    else:
        if prevstate['momos']:
            sum-=60
            prevstate['momos']=False
    if burger.get()==65:
        if not prevstate['burger']:
            sum+=65
            prevstate['burger']=True
        
    else:
        if prevstate['burger']:
            sum-=65
            prevstate['burger']=False
    if samosa.get()==20:
        if not prevstate['samosa']:
            sum+=20
            prevstate['samosa']=True

    else:
        if prevstate['samosa']:
            sum-=20
            prevstate['samosa']=False
    if cd.get()==25:
        if not prevstate['cd']:
            sum+=25
            prevstate['cd']=True

    else:
        if prevstate['cd']:
            sum-=25
            prevstate['cd']=False
    totamt.config(text=f"Total Amount : {sum}")


def generate_bill():
    bill_text = "Bill:\n"
    total=0
    if coffee.get() == 30:
        bill_text += "Coffee: Rs.30\n"
        total += 30
    if tea.get() == 20:
        bill_text += "Tea: Rs.20\n"
        total += 20
    if momos.get() == 60:
        bill_text += "Momos: Rs.60\n"
        total += 60
    if burger.get() == 65:
        bill_text += "Burger: Rs.65\n"
        total += 65
    if samosa.get() == 20:
        bill_text += "Samosa: Rs.20\n"
        total += 20
    if cd.get() == 25:
        bill_text += "Cold drink: Rs.25\n"
        total += 25
    print(bill_text)
    print(f"Total Amount : Rs.{total}")
    

checkbox1=Checkbutton(root,text="Coffee - Rs.30",variable=coffee,onvalue=30,offvalue=0,command=show)
checkbox2=Checkbutton(root,text="Tea - Rs.20",variable=tea,onvalue=20,offvalue=0,command=show)
checkbox3=Checkbutton(root,text="Momos - Rs.60",variable=momos,onvalue=60,offvalue=0,command=show)
checkbox4=Checkbutton(root,text="Burger - Rs.65",variable=burger,onvalue=65,offvalue=0,command=show)
checkbox5=Checkbutton(root,text="Samosa - Rs.20",variable=samosa,onvalue=20,offvalue=0,command=show)
checkbox6=Checkbutton(root,text="Cold drink - Rs.25",variable=cd,onvalue=25,offvalue=0,command=show)
bill=Button(root,text="Generate Bill",bg='green',fg='white',command=generate_bill)

totamt=Label(text="Total Amount : ",bg='white')
heading.grid(row=0,column=0,columnspan=3,ipadx=200)
sub_head.grid(row=1,column=0,columnspan=3)
checkbox1.grid(row=2,column=0,padx=10,pady=40)
checkbox2.grid(row=2,column=1,padx=10,pady=40)
checkbox3.grid(row=3,column=0,padx=10,pady=40)
checkbox4.grid(row=3,column=1,padx=10,pady=40)
checkbox5.grid(row=4,column=0,padx=10,pady=40)
checkbox6.grid(row=4,column=1,padx=10,pady=40)
totamt.grid(row=5,column=0,padx=10,pady=40,ipadx=15,ipady=15)
bill.grid(row=5,column=1,padx=10,pady=40,ipadx=15,ipady=15)

root.mainloop()
```
